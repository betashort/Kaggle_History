# rating
ratingは、10段階の評価値になっている。  
-1は、見たけど評価していない


```python
#rating['rating'].name = "user_rating"
user.rename(columns={'rating': 'user_rating'}, inplace=True)
```

```python
#min
max_user_rating = user.groupby("user_id")["user_rating"].max().reset_index()
max_user_rating.rename(columns={"user_rating": "max_user_rating"}, inplace=True)

#max
min_user_rating = user.groupby("user_id")["user_rating"].min().reset_index()
min_user_rating.rename(columns={"user_rating": "min_user_rating"}, inplace=True)

#mean
mean_user_rating = user.groupby("user_id")["user_rating"].mean().reset_index()
mean_user_rating.rename(columns={"user_rating": "mean_user_rating"}, inplace=True)

#median
median_user_rating = user.groupby("user_id")["user_rating"].median().reset_index()
median_user_rating.rename(columns={"user_rating": "median_user_rating"}, inplace=True)

#std
std_user_rating = user.groupby("user_id")["user_rating"].std().reset_index()
std_user_rating.rename(columns={"user_rating": "std_user_rating"}, inplace=True)
```

```python
user = pd.merge(user, min_user_rating, on="user_id", how="left")
user = pd.merge(user, max_user_rating, on="user_id", how="left")
user = pd.merge(user, mean_user_rating, on="user_id", how="left")
user = pd.merge(user, median_user_rating, on="user_id", how="left")
user = pd.merge(user, std_user_rating, on="user_id", how="left")
```

```python
user["relative_user_rating"] = (user["user_rating"]-user["min_user_rating"])/(user["max_user_rating"]-user["min_user_rating"])
user["relative_user_rating"].fillna(0.5, inplace=True)
```
